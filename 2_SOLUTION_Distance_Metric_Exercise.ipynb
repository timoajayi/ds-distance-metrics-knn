# Solution: Distance metric exercise

This is only one possible implementation. It's totally fine if your solution looks different as long as it works ;) 

```python
import numpy as np


def distance(pt1, pt2, c=2, verbose=True):
    metrics = ['Manhattan', 'Euclidean', 'Minkowski']
    if verbose:
        if c<3:
            print('This distance metric is a measurement of {}'.format(metrics[c-1]))
        else:
            print('This distance metric is a measurement of Minkowski with order {}'.format(c))
    dist = np.power(sum([np.abs(xi-yi)**c for xi, yi in zip(pt1, pt2)]),1/c)
    return dist

pt1 = (3, 5)
pt2 = (1, 9)
print(distance(pt1, pt2))
print(distance(pt1, pt2, c=1)) 
print(distance(pt1, pt2, c=3)) 
print(distance(pt1, pt2, c=5))
```
